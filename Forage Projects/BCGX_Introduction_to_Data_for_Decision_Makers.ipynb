# Campaign Performance Analysis using Python:
Analyze the Campaign dataset using Pandas pivot tables to summarize the total sales by channel and by campaign, and present the results.

## Load the data
Import the pandas library and load the dataset into a DataFrame, then display the head and info of the DataFrame to understand its structure and content.


```python
# import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# file location in drive
file_path = '/content/drive/MyDrive/Forage Virtual Internships/BCG X/Campaign_Data.xlsx'

# Load data into pandas dataframe
df = pd.read_excel(file_path)

# Display the first five rows
display(df.head())

# Display dataset info
display(df.info())
```

## Analyze total sales by channel
Use a pandas pivot table to calculate the total sales for each channel and display the result.


```python
# Calculate total sales by channel
sales_by_channel = pd.pivot_table(df, values='Sales ($)', index='Channel', aggfunc='sum')
print("Total Sales by Channel:")
display(sales_by_channel)
```

## Analyze total sales by campaign
Use a pandas pivot table to calculate the total sales for each campaign type.


```python
# Calculate total sales by campaign
sales_by_campaign = pd.pivot_table(df, values='Sales ($)', index='Campaign Type', aggfunc='sum')
print("\nTotal Sales by Campaign:")
display(sales_by_campaign)
```

## Summary of Findings

Based on the analysis of the campaign data:

**Total Sales by Channel:**

*   **Email:** Generated the highest total sales.
*   **Instagram:** Generated the second highest total sales.
*   **Web Banner:** Generated the lowest total sales.

**Total Sales by Campaign:**

*   **Campaign B (Sales-focused):** Generated significantly higher total sales compared to Campaign A.
*   **Campaign A (Informal):** Generated lower total sales compared to Campaign B.

**Key Takeaways:**

*   Email is the most effective channel in terms of overall revenue generation.
*   Campaign B's sales-focused messaging resulted in higher overall sales.

**Recommendation:**

To maximize overall revenue, NewCo should consider focusing more resources on **Email** as a channel and utilizing **Campaign B's** sales-focused messaging. Further analysis could explore the performance of specific channel and campaign combinations and the impact on new customer acquisition.

## Analyze Total Sales by Channel and Campaign with Customer Type Filter

```python
# Calculate total sales by channel, campaign with customer type filter
sales_by_channel_campaign_customer = pd.pivot_table(df, values='Sales ($)', index=['Channel', 'Campaign Type'], columns='Customer Type', aggfunc='sum')
display(sales_by_channel_campaign_customer)
```

## Summary of Findings by Channel, Campaign, and Customer Type

Based on the pivot table analyzing sales by channel, campaign type, and customer type:

*   **New Customers:**
    *   **Email + Campaign A** generated the highest sales among new customers.
    *   **Email + Campaign B** and **Instagram + Campaign A** also performed relatively well with new customers.
    *   Web Banner generally had lower sales with new customers compared to Email and Instagram.
*   **Existing Customers:**
    *   **Email + Campaign B** generated the highest sales among existing customers, significantly higher than other combinations.
    *   **Instagram + Campaign A and B**, and **Web Banner + Campaign B** also contributed to sales from existing customers, but to a lesser extent than Email + Campaign B.

**Key Takeaways:**

*   For acquiring **new customers**, **Email with Campaign A's informal tone** appears to be the most effective combination.
*   For maximizing sales from **existing customers**, **Email with Campaign B's salesy tone** is the strongest performer.
*   Different channel and campaign combinations have varying effectiveness depending on whether the target is a new or existing customer.

**Recommendation:**

NewCo should consider tailoring its marketing strategy based on the customer type they are targeting. To acquire **new customers**, focus on **Email with Campaign A**. To maximize sales from **existing customers**, prioritize **Email with Campaign B**. Further analysis could involve visualizing these findings for easier comparison.

## Visualizing Performance for New Customers

### 1. Sales from New Customers by Channel

```python
# Filter for new customers
new_customers_df = df[df['Customer Type'] == 'New']

# Calculate total sales by channel for new customers
sales_by_channel_new = pd.pivot_table(new_customers_df, values='Sales ($)', index='Channel', aggfunc='sum')

# Create a bar plot
plt.figure(figsize=(8, 5))
sns.barplot(data=sales_by_channel_new, x=sales_by_channel_new.index, y='Sales ($)')
plt.title('Sales from New Customers by Channel')
plt.xlabel('Channel')
plt.ylabel('Total Sales ($)')
plt.show()
```

### 2. Sales from New Customers by Campaign

```python
# Filter for new customers
new_customers_df = df[df['Customer Type'] == 'New']

# Calculate total sales by campaign for new customers
sales_by_campaign_new = pd.pivot_table(new_customers_df, values='Sales ($)', index='Campaign Type', aggfunc='sum')

# Create a bar plot
plt.figure(figsize=(8, 5))
sns.barplot(data=sales_by_campaign_new, x=sales_by_campaign_new.index, y='Sales ($)')
plt.title('Sales from New Customers by Campaign')
plt.xlabel('Campaign Type')
plt.ylabel('Total Sales ($)')
plt.show()
```

### 3. Sales from New Customers by Channel and Campaign

```python
# Filter for new customers
new_customers_df = df[df['Customer Type'] == 'New']

# Calculate total sales by channel and campaign for new customers
sales_by_channel_campaign_new = pd.pivot_table(new_customers_df, values='Sales ($)', index=['Channel', 'Campaign Type'], aggfunc='sum').unstack()

# Create a grouped bar plot
sales_by_channel_campaign_new.plot(kind='bar', figsize=(10, 6))
plt.title('Sales from New Customers by Channel and Campaign')
plt.xlabel('Channel and Campaign Type')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=0)
plt.legend(title='Campaign Type')
plt.tight_layout()
plt.show()
```

```python
# Filter the pivot table to focus on new customers
new_customer_sales = sales_by_channel_campaign_customer['New'].unstack()

# Create a bar plot for new customer sales by channel and campaign
plt.figure(figsize=(10, 6))
sns.barplot(data=new_customer_sales, x=new_customer_sales.index, y='A', color='skyblue', label='Campaign A')
sns.barplot(data=new_customer_sales, x=new_customer_sales.index, y='B', color='lightcoral', label='Campaign B')

plt.title('Sales from New Customers by Channel and Campaign')
plt.xlabel('Channel')
plt.ylabel('Total Sales ($)')
plt.legend(title='Campaign Type')
plt.show()
```

## Key Takeaways from Visualizations

Based on the charts showing sales from new customers:

*   **Sales by Channel:** Email is the most effective channel for generating sales from new customers, followed by Instagram and then Web Banner.
*   **Sales by Campaign:** Campaign A (informal tone) is more effective than Campaign B (salesy tone) in driving sales from new customers.
*   **Sales by Channel and Campaign:** The combination of **Email and Campaign A** is clearly the strongest performing combination for attracting new customers and generating sales. Email + Campaign B is the second strongest combination.

These visualizations support the earlier pivot table findings and highlight the importance of using the right channel and messaging for new customer acquisition.
