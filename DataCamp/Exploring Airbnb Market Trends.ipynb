As a consultant working for a real estate start-up, you have collected Airbnb listing data from various sources to investigate the short-term rental market in New York. You'll analyze this data to provide insights on private rooms to the real estate company.

There are three files in the data folder: **`airbnb_price.csv`**, **`airbnb_room_type.xlsx`**, **`airbnb_last_review.tsv`**.

- What are the dates of the earliest and most recent reviews? Store these values as two separate variables with your preferred names.
- How many of the listings are private rooms? Save this into any variable.
- What is the average listing price? Round to the nearest two decimal places and save into a variable.
- Combine the new variables into one DataFrame called **`review_dates`** with four columns in the following order: **`first_reviewed`**, **`last_reviewed`**, **`nb_private_rooms`**, and **`avg_price`**. The DataFrame should only contain one row of values.

```python
# Import necessary packages
import pandas as pd
import numpy as np

# Loading the data using pandas methods to read files
airbnb_price = pd.read_csv('data/airbnb_price.csv')
airbnb_room_type = pd.read_excel('data/airbnb_room_type.xlsx')
airbnb_last_review = pd.read_csv('data/airbnb_last_review.tsv', sep = '\t')

# Preview data using .head() method
print(airbnb_price.head(), '\n')
print(airbnb_room_type.head(), '\n')
print(airbnb_last_review.head(), '\n')

# Merging the three DataFrames using pd.merge() function 
merge1 = airbnb_price.merge(airbnb_room_type, on = 'listing_id')
merge_airbnb_dfs = merge1.merge(airbnb_last_review, on = 'listing_id')
merge_airbnb_dfs.head()
```

```python
# Display information about merge_airbnb_dfs
merge_airbnb_dfs.info()
```

- ### What are the dates of the earliest and most recent reviews? Store these values as two separate variables with your preferred names.

```python
# convert the reviews column to a date format.
merge_airbnb_dfs['last_review'] = \
pd.to_datetime(merge_airbnb_dfs['last_review'], format = '%B %d %Y', origin = 'unix')


# Determining the earliest and most recent review dates.
earliest_review_date = merge_airbnb_dfs['last_review'].min()
recent_review_date = merge_airbnb_dfs['last_review'].max()
```

- ### How many of the listings are private rooms? Save this into any variable.

```python
# Display all unique values in the room_type column.
merge_airbnb_dfs['room_type'].unique()
```

### Observation(s):
The **`room_type`** column has variations in capitalization as shown in the output above. Hence, there's need to standardize the cases to be in all lower or upper case.

```python
# Convert all values in room_type column to lower case using the .str constructor and the .lower() method.
merge_airbnb_dfs['room_type'] = merge_airbnb_dfs['room_type'].str.lower()
merge_airbnb_dfs['room_type'].unique()
```

```python
# Subset the dataframe to select room_type == 'private room' and use the shape attribute to obtain the count.
private_room_count = \
merge_airbnb_dfs[merge_airbnb_dfs['room_type'] == 'private room'].shape[0]
private_room_count
```

- ### What is the average listing price? Round to the nearest two decimal places and save into a variable.

```python
# Display the first five values of the price column
merge_airbnb_dfs['price'].head()
```

### Observation(s):
The **`price`** column contains **'dollars'** in its values which makes it a string type variable. This column requires cleaning to enable us calculate the average price of **'listings'**

```python
# Remove " dollars" from price using .str.replace on it.
merge_airbnb_dfs['price'] = merge_airbnb_dfs['price'].str.replace(" dollars", '')

# Convert the data type of price from object to float.
merge_airbnb_dfs['price'] = merge_airbnb_dfs['price'].astype(float)
print(merge_airbnb_dfs['price'].head())
```

```python
# Calculate average listing price.
ave_listing_price = merge_airbnb_dfs['price'].mean().round(2)
ave_listing_price
```

- ### Combine the new variables into one DataFrame called **`review_dates`** with four columns in the following order: **`first_reviewed`**, **`last_reviewed`**, **`nb_private_rooms`**, and **`avg_price`**. The DataFrame should only contain one row of values.

```python
# Create a dictionary of first_reviewed, last_reviewed, nb_private_rooms and avg_price with their respective values.
dict_combined = {'first_reviewed': [earliest_review_date],
                 'last_reviewed': [recent_review_date],
                 'nb_private_rooms': [private_room_count],
                 'avg_price': [ave_listing_price]}

# Pass a dictionary into the pd.DataFrame() constructor.
review_dates = pd.DataFrame(dict_combined, index=[0])
review_dates
```
